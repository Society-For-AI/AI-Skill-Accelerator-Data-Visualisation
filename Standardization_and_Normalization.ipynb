### Normalisation Vs. Standardization - Feature Scaling

Feature Scaling is an important step to take prior to training machine learning models to ensure that features are within the same scale.

Example: Interest rate and employment score are at a different scale. This will result in one feature dominating the other feature.

```python
# Import dataset and libraries

import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Define the ranges
interest_rate_min, interest_rate_max = 1.814688, 2.258229
employment_min, employment_max = 49.908353, 63.668422
sp500_min, sp500_max = 2078.553895, 2486.474488

# Generate 20 random values within the specified ranges
interest_rates = np.random.uniform(interest_rate_min, interest_rate_max, 20)
employment = np.random.uniform(employment_min, employment_max, 20)
sp500_prices = np.random.uniform(sp500_min, sp500_max, 20)

# Create the DataFrame
data = {
    'Interest_Rates': interest_rates,
    'Employment': employment,
    'SP500_Price': sp500_prices
}

stock_df = pd.DataFrame(data)
```

```python
stock_df
```

```python
sns.pairplot(stock_df)
```

```python
stock_df.describe().round(2)
```

### Normalization - Scaling From 0 to 1

Normalization is conducted to make values range from 0 to 1

```python
stock_df
```

```python
stock_df.describe()
```

```python
stock_df['SP500_Price'].min()
```

```python
stock_df['SP500_Price'].max()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
stock_df_scaled = scaler.fit_transform(stock_df)
```

```python
stock_df_scaled = pd.DataFrame(stock_df_scaled, columns = ['Interest Rates', 'Employment', 'S&P 500 Price'])
```

```python
sns.pairplot(stock_df_scaled)
```

```python
stock_df_scaled.describe().round(2)
```

### Standardization



*   Standardization is conducted to transform the data to have a mean of zero and standard deviation of 1.
*   Standardization is also known as Z-score normalization.
*   Standardization is preferred over normalization when there are a lot of outliers.





```python
stock_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
stock_df_scaled2 = scaler.fit_transform(stock_df)
```

```python
stock_df_scaled2 = pd.DataFrame(stock_df_scaled2, columns = ['Interest Rates', 'Employment', 'S&P 500 Price'])
```

```python
sns.pairplot(stock_df_scaled2)
```
